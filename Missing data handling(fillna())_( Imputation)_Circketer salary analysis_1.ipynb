# 2nd methord from sklearn SimpleImputation

```python
Data=pd.read_csv('cricket_salary_data.csv')
Data.head(13)
```

```python
feature=Data.iloc[:,0:3]
feature
```

```python
label=Data.iloc[:,3]
label
```

```python
from sklearn.impute import SimpleImputer
from numpy import nan
```

```python
imputer=SimpleImputer(missing_values=nan,strategy='mean')
imputer
```

```python
feature.iloc[:,1:2]=imputer.fit_transform(feature.iloc[:,1:2])
feature.iloc[:,1:2]
```
